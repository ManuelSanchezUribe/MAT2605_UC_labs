### ¿Qué es Jupyter Notebook?
Es un entorno de trabajo interactivo que permite intercalar bloques de texto (como este, escrito en Markdown), ecuaciones, visualizaciones y bloques de código ejecutable. Anaconda ya incluye por defecto Jupyter Notebook y las librerías de computación científica (NumPy, SciPy, Matplotlib, Pandas), por lo que estamos listos para comenzar.

### 1. Variables y Tipos de Datos
En Python, no es necesario declarar el tipo de variable explícitamente. El lenguaje lo infiere dinámicamente en el momento de la asignación.

```python
# Variables y tipos de datos elementales en Python
entero = 10          # int (Integer)
decimal = 3.14159    # float (Punto flotante)
texto = "Hola MAT2605"   # str (String o cadena de texto)
booleano = True      # bool (Booleano: True o False)

# Imprimir variables y sus tipos
print(f"El valor es {decimal} y su tipo es {type(decimal)}")
print(f"El valor es {texto} y su tipo es {type(texto)}")
```

### 2. Control de Flujo (If, Elif, Else)
Las sentencias condicionales permiten ejecutar distintos bloques de código dependiendo de si se cumple o no una condición. En Python, la indentación (los espacios al inicio de la línea) define los bloques de código.

```python
# Control de flujo básico
valor = 5

if valor > 0:
    print("El valor es estrictamente positivo")
elif valor == 0:
    print("El valor es exactamente cero")
else:
    print("El valor es negativo")
```

### 3. Ciclos (For loops)
Los ciclos `for` en Python se utilizan para iterar sobre una secuencia (como una lista, una tupla o un rango de números).

```python
# Iterando sobre una lista
mi_lista = [2, 4, 6, 8]
print("Elementos de la lista:")
for numero in mi_lista:
    print(numero)

print("\nIterando con la función range():")
# range(inicio, fin, paso) - el 'fin' no se incluye
for i in range(0, 5): 
    print(f"Iteración número {i}")
```

### 4. Definición de Funciones
Las funciones permiten empaquetar y reutilizar código. Se definen utilizando la palabra reservada `def`.

```python
# Creación de una función matemática simple
def evaluar_signo(numero):
    """Esta función recibe un número y retorna una cadena de texto indicando su signo."""
    if numero > 0:
        return "Positivo"
    elif numero < 0:
        return "Negativo"
    else:
        return "Cero"

# Llamamos a la función
resultado = evaluar_signo(-12.5)
print("El número ingresado es:", resultado)

# Lambda function
funlambda = lambda x: x**2+x -1
funlambda(0.1)
```

Python base no está optimizado para operaciones matemáticas intensivas. Para el cálculo científico y análisis numérico, utilizamos **NumPy** (operaciones vectorizadas y matriciales) y **Matplotlib** (visualización de datos).

### 6. Visualización de Datos
Utilizaremos `matplotlib.pyplot` para generar gráficos de manera rápida.

```python
# Crear un vector de valores espaciados uniformemente
# np.linspace(inicio, fin, número de puntos)
x = np.linspace(-10, 10, 100)

# Aplicar funciones matemáticas vectorizadas
y1 = np.sin(x)
y2 = np.cos(x)

# Generar el gráfico
plt.figure(figsize=(8, 4)) # Tamaño de la figura
plt.plot(x, y1, label="sen(x)", color="blue", linewidth=2)
plt.plot(x, y2, label="cos(x)", color="red", linestyle="--")

plt.title("Gráfico de funciones trigonométricas")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.legend() # Mostrar las etiquetas (labels)

# Mostrar el gráfico
plt.show()
```
