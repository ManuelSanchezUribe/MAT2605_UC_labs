    <h3 style="color: #d35400; margin-top: 0;">📚 Teoría:</h3>
    <p>Norma vectorial (norma p): $\|x\|_p = \left( \sum_{j=1}^N |x_j|^p \right)^{1/p}, \quad p\in[1,\infty), \quad x\in\mathbb{R}^N, \qquad \text{norma infinito }\|x\|_\infty = \max_{1\leq i\leq N} |x_i|$
    </p>
    Norma matricial inducida: $\|A\| = \max_{x\in\mathbb{R}^N,\,\|x\|=1} \frac{\|Ax\|}{\|x\|}, \quad A\in \mathbb{R}^{N\times N}$
    <p>Norma Frobenius $\|A\|_F = \sqrt{\sum_{i,j=1}^N |a_{ij}|^2}$
    <p>N&uacute;mero de condici&oacute;n</p>
    $$\mathrm{cond}(A) = \|A\|\,\|A^{-1}\|$$
    <p>M&eacute;todo iterativo b&aacute;sico</p>
    $$ x^{(k+1)} = T x^{(k)} + b, \quad x^{(0)} \text{ vector inicial dado}$$
    donde $T$ es la matriz de iteraci&oacute;n. (Para asegurar convergencia es suficiente que $\rho(T)<1$ donde $\rho$ es el radio espectral. Se analiza en detalles durante las clases de la semana que viene).

# Normas

En Python (NumPy) es posible calcular normas mediante <code>np.linalg.norm(x)</code> que calcula la norma euclidiana del vector $x$.

Para calcular las normas $\|\cdot\|_p$ se puede usar un segundo argumento: <code>np.linalg.norm(x,p)</code>.
Para la norma infinito se usa el "s&iacute;mbolo" $\infty$ que se representa en Python (NumPy) como <code>np.inf</code>.

En Python es posible calcular la norma $p$ para $p=1,2,\infty$ de una matriz, pero no es posible con otros valores de $p$.
Es posible calcular la norma Frobenius con <code>np.linalg.norm(A)</code> o <code>np.linalg.norm(A,'fro')</code>

*Cuidado*: Para calcular la norma inducida por la norma $2$ hay que usar <code>np.linalg.norm(A,2)</code>

```python
# Normas vectoriales
x = np.array([1,1,1,1])
print("Vector x: ",x)

print("Norma euclidiana = ", np.linalg.norm(x))
print("Norma 1 = ", np.linalg.norm(x,1))
print("Norma inf = ",np.linalg.norm(x,np.inf))

print("Norma Pi = ",np.linalg.norm(x,np.pi))

# Normas matriciales
A = np.array([[1,-1,-1,-1],[0,1,-1,-2],[0,0,1,0],[0,0,0,3]])
print("Matriz A:\n",A)

print("Norma 2 = ", np.linalg.norm(A,2))
print("Norma 1 = ", np.linalg.norm(A,1))
print("Norma inf = ",np.linalg.norm(A,np.inf))
print("Norma Frobenius = ",np.linalg.norm(A,'fro'))
print("Norma (sin dar segundo argumento) = ",np.linalg.norm(A))

print("Norma Pi = ",np.linalg.norm(A,np.pi))
```

# N&uacute;mero de condici&oacute;n

En Python (NumPy) podemos usar <code>np.linalg.cond(A,arg)</code> para calcular el n&uacute;mero de condici&oacute;n de una matriz $A$. (<code>arg</code> se elige como antes para las normas)

*Cuidado:* <code>np.linalg.cond(A)</code> calcula <code>np.linalg.cond(A,2)</code>

```python
print('Numero condicion (norma Frobenius) = ',np.linalg.cond(A,'fro'))
print('Numero condicion (norma 2) = ',np.linalg.cond(A,2))
print('Numero condicion (norma 1) = ',np.linalg.cond(A,1))
print('Numero condicion (norma 1) = ',np.linalg.cond(A,np.inf))
```

Tenemos la cota
$$\frac{\|\Delta x\|}{\|x\|}\leq \mathrm{cond}(A) \frac{\|\Delta b\|}{\|b\|}$$

En el c&oacute;digo abajo se usa la norma $2$. Que se observa cuando se cambia a la norma $1$ y/o $\infty$?

```python
delta = 0.00001
p = 2
A = np.array([[1,1+delta],[1-delta,1]])
print('Matrix A = \n',A)
condA = np.linalg.cond(A,p)

# resolver systema lineal
b = np.array([1,1])
delta_b = np.array([-1,1])/100
x = np.linalg.solve(A,b)
delta_x = np.linalg.solve(A,delta_b)

# errores
err_rel_x = np.linalg.norm(delta_x,p)/np.linalg.norm(x,p)
err_rel_b = np.linalg.norm(delta_b,p)/np.linalg.norm(b,p)
print('error rel. en solucion = ',err_rel_x)
print('error rel. en dato b   = ',err_rel_b)
print('amplificacion          = ',err_rel_x/err_rel_b)
print('numero de condicion    = ',condA)

# otro ejemplo con perturbacion aleatoria
# resolver systema lineal
print('\n\nOtro ejemplo\n')
delta_b = np.random.rand(2)/1000
x = np.linalg.solve(A,b)
delta_x = np.linalg.solve(A,delta_b)

# errores
err_rel_x = np.linalg.norm(delta_x,p)/np.linalg.norm(x,p)
err_rel_b = np.linalg.norm(delta_b,p)/np.linalg.norm(b,p)
print('error rel. en solucion = ',err_rel_x)
print('error rel. en dato b   = ',err_rel_b)
print('amplificacion          = ',err_rel_x/err_rel_b)
print('numero de condicion    = ',condA)
```

# M&eacute;todo iterativo b&aacute;sico

El m&eacute;todo de Jacobi se define como
$$x^{(k+1)} = T_J x^{(k)} + c, \quad k=0,1,2,...$$
donde $T_J = D^{-1}(L+U)$, con $A = -L + D -U$ y $c = D^{-1}b$.

Es importante en una implementaci&oacute;n asegurarse que termine en pasos finitos. 
En el c&oacute;digo abajo se acota el numero de iteraciones con <code>max_iter</code>

Ademas, se incluye el criterio de parada: <code>norm_residual < tol</code>
que se traduce a: Si 
$$\|Ax^{(k)}-b\|_2< \mathrm{tol}$$
entonces termina. 
Aqui, $\|Ax^{(k)}-b\|_2$ es la norma del residuo en el $k$-esimo paso. 

Muchas veces se usa el criterio
$$\|Ax^{(k)}-b\|_2< \mathrm{tol} \times \|Ax^{(0)}-b\|_2$$
donde $\|Ax^{(0)}-b\|_2$ es la norma del residuo inicial. 
Modifique el c&oacute;digo abajo para incluir este criterio.

## Jacobi

```python
def metodo_de_Jacobi(A, b, x0, max_iter=100, tol=1e-8):
    """
    Resuelve el sistema Ax = b usando el método de Jacobi.
    Input: A, b, x0 (vector inicial)
    Output: x (solución aproximada), residuals (historial de errores)
    """
    x = x0.copy()
    n = len(b)
    residuals = []
    
    for k in range(max_iter):
        # Calcular el residuo actual
        residual = b - np.dot(A, x)
        norm_residual = np.linalg.norm(residual)
        residuals.append(norm_residual)
        
        # Criterio de parada
        if norm_residual < tol:
            print(f"J: Convergencia alcanzada en la iteración {k+1} | Error: {norm_residual:.6e}")
            return x, residuals
        x_old = x.copy()
        # Calcular los nuevos valores de x
        for i in range(n):
            suma = np.dot(A[i, :i], x_old[:i]) + np.dot(A[i, i+1:], x_old[i+1:])
            # Jacobi
            x[i] = (b[i] - suma) / A[i, i]
            
    print(f"J: Número máximo de iteraciones alcanzado. Error final: {norm_residual:.6e}")
    return x, residuals
```

```python
A1 = np.array([[7,-4],[-8,9]])
b1 = np.array([3,-4])
x0 = np.array([-1.0,3.0])
x = metodo_de_Jacobi(A1,b1,x0)
```

```python
A2 = np.array([[1.0, 0.5,0.5],[0.5,1.0,0.5],[0.5,0.5,1.0]])
b2 = np.array([1.0,-.1,0.0])
x = metodo_de_Jacobi(A2,b2,np.ones(3))
```

## An&aacute;lisis con las matrices de iteraci&oacute;n

```python
D = np.diag(np.diag(A2))
L = -np.tril(A2, k=-1)
U = -np.triu(A2, k=1)
print(" D:\n", D,"\n L:\n", L,"\n U:\n", U)

# Matriz e iteracion de Jacobi
T_J = np.linalg.inv(D)@ (L+U)
print("T_J:\n",T_J)

# condicion de radio espectral
rho_J = max(abs(np.linalg.eigvals(T_J)))
print(" rho_J = ", rho_J, "==> no converge!")
```

```python
A3 = np.array([[1,2,-2],[1,1,1],[2,2,1]],dtype = np.float64)
D = np.diag(np.diag(A3))
L = np.tril(A3, k=-1)
U = np.triu(A3, k=1)

# Matriz e iteracion de Jacobi
T_J = -np.linalg.inv(D)@ (L+U)

print("T_J:\n",T_J)
# condicion de radio espectral
rho_J = max(abs(np.linalg.eigvals(T_J)))
print(" rho_J = ", rho_J, "==> converge!")
```
